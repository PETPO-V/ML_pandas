**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
%%time

def mean_reading_score_by_race(df):
    scores_by_race = {}
    race_column = df['race/ethnicity']
    reading_scores = df['reading score']

    for race, score in zip(race_column, reading_scores):
        if race not in scores_by_race:
            scores_by_race[race] = [score]
        else:
            scores_by_race[race].append(score)

    mean_scores = {}
    for race, scores in scores_by_race.items():
        mean_scores[race] = sum(scores) / len(scores)

    return mean_scores

def min_writing_score_by_parental_education(df):
    scores_by_education = {}
    education_column = df['parental level of education']
    writing_scores = df['writing score']

    for education, score in zip(education_column, writing_scores):
        if education not in scores_by_education:
            scores_by_education[education] = [score]
        else:
            scores_by_education[education].append(score)

    min_scores = {}
    for education, scores in scores_by_education.items():
        min_scores[education] = min(scores)

    return min_scores

mean_reading_scores = mean_reading_score_by_race(df)
min_writing_scores = min_writing_score_by_parental_education(df)

print("Average reading score by group:")
for race, score in mean_reading_scores.items():
    print(f"{race}: {score}")

print("\nThe minimum score for a writing according to the level of education:")
for education, score in min_writing_scores.items():
    print(f"{education}: {score}")
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return 'A'
    elif average_mark >= 80:
        return 'B'
    elif average_mark >= 70:
        return 'C'
    elif average_mark >= 60:
        return 'D'
    elif average_mark >= 50:
        return 'E'
    else:
        return 'F'


df['Average'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
df['Grade'] = df['Average'].apply(GetGrade)
grade_counts = df['Grade'].value_counts().sort_index(ascending=False)

display(df[['math score', 'reading score', 'writing score', 'Average', 'Grade']])
print(grade_counts)
```
